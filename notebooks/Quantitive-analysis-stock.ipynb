## Load the data

```python
# Load the data 

aapl_data = pd.read_csv('../data/AAPL_historical_data.csv')
amzn_data = pd.read_csv('../data/AMZN_historical_data.csv')
goog_data = pd.read_csv('../data/GOOG_historical_data.csv')
meta_data = pd.read_csv('../data/META_historical_data.csv')
msft_data = pd.read_csv('../data/MSFT_historical_data.csv')
nvda_data = pd.read_csv('../data/NVDA_historical_data.csv')
tsla_data = pd.read_csv('../data/TSLA_historical_data.csv')
```

## Checking the required columns are present

```python
required_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']

for df in [aapl_data, amzn_data, goog_data, meta_data, msft_data, nvda_data, tsla_data]:
    if not all(col in df.columns for col in required_columns):
        print("Missing columns in DataFrame")
    else:
        print("All required columns are present.")
```

```python
# Combine the data in to stocl_data
stock_data = {
    'AAPL': aapl_data,
    'AMZN': amzn_data,
    'GOOG': goog_data,
    'META': meta_data,
    'MSFT': msft_data,
    'NVDA': nvda_data,
    'TSLA': tsla_data
}
```

```python
# Concatenate the adjusted close prices into a single DataFrame
adj_close_data = pd.concat([df['Adj Close'].rename(ticker) for ticker, df in stock_data.items()], axis=1)
```

## Calculate the SMA

```python
import talib

# Calculate the 50-day and 200-day SMA for each stock
for ticker, df in stock_data.items():
    df[f'{ticker}_SMA_50'] = talib.SMA(df['Adj Close'], timeperiod=50)
    df[f'{ticker}_SMA_200'] = talib.SMA(df['Adj Close'], timeperiod=200)
```

## Calculate Volatility

```python
# Calculate daily returns
aapl_data['Returns'] = aapl_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
aapl_data['Volatility'] = aapl_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#..
amzn_data['Returns'] = amzn_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
amzn_data['Volatility'] = amzn_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#
goog_data['Returns'] = goog_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
goog_data['Volatility'] = goog_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#
meta_data['Returns'] = meta_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
meta_data['Volatility'] = meta_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#
msft_data['Returns'] = msft_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
msft_data['Volatility'] = msft_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#
nvda_data['Returns'] = nvda_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
nvda_data['Volatility'] = nvda_data['Returns'].rolling(window=30).std() * np.sqrt(252)
#
tsla_data['Returns'] = tsla_data['Adj Close'].pct_change()

# Calculate rolling volatility ( 30-day volatility)
tsla_data['Volatility'] = tsla_data['Returns'].rolling(window=30).std() * np.sqrt(252)
# Display the first few rows
# print(aapl_data[['Volatility']].head())
```
